```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
from matplotlib.ticker import MultipleLocator
import seaborn as sns
%matplotlib inline

#df = data frame, 2 dimensional data structure or table with rows and columns
df = pd.read_csv('FormResponses1.csv')

sns.set_theme(style="whitegrid")

#Histogram 1: Usage Intensity
#To demonstrate how much
plt.figure(figsize=(10, 8))
ax = sns.histplot(data=df, x='On average how many minutes per day do you spend chatting with an AI? ', 
             bins=17, kde=True,
             color='royalblue')

ax.yaxis.set_major_locator(MultipleLocator(5))

plt.title('Daily AI Usage Time', fontsize=14)
plt.xlabel('Minutes per Day')
plt.show()
```

```python
#Histogram 2: Politeness vs. Sincerity 
#Deception aspect
plt.figure(figsize=(10, 6))

# 2. Plotting the first variable (Politeness)
sns.histplot(df['I feel a moral obligation be polite or kind when messaging the AI.'], 
             color="royalblue", label="Politeness Rating", kde=True, discrete=True, alpha=0.6)

# 3. Plotting the second variable (Sincerity)
sns.histplot(df['When the AI claims to understand my feelings, I perceive that sentiment as sincere.'], 
             color="orange", label="Sincerity Rating", kde=True, discrete=True, alpha=0.7)

# 4. Customizing the X-Axis for 1-5 ratings
plt.gca().xaxis.set_major_locator(MultipleLocator(1))
plt.gca().yaxis.set_major_locator(MultipleLocator(2))

# 5. Legend and Labels
plt.title('Needing to be Polite Towards AI vs. Is AI Sincere', fontsize=14)
plt.xlabel('Rating (1 = Strongly Disagree, 5 = Strongly Agree)')
plt.ylabel('Number of Responses')
plt.legend()
plt.show()
```

```python
#Histogram 3: Sense of Loss (Emotional Dependency)
plt.figure(figsize=(10, 6))

#  plot
ax3 = sns.histplot(data=df, x='I would feel a significant sense of loss if I could no longer access my preferred AI.', 
                  bins=5, color='orchid', kde=True, discrete=True, kde_kws={'bw_adjust':0.65})

# Multiples of 2 on Y-axis and 1 on X-axis
ax3.set_yticks(range(0, 22, 2))
ax3.set_xticks([1, 2, 3, 4, 5])

# Labels
plt.title('Emotional Dependency: Distribution of "Sense of Loss"', fontsize=14)
plt.xlabel('Rating (1 = No Loss, 5 = Significant Loss)')
plt.ylabel('Number of Respondents')
plt.show()
```
